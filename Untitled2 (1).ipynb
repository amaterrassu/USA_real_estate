```python
def custom_hist(df : pd.Series):
  
  figure = plt.figure(figsize = (7, 7))
  ax = figure.add_subplot()
  plt.hist(df, color = 'red', edgecolor = 'skyblue', bins = 20, density = True, label = 'hist')
  plt.axvline(df.mean(), color='black', linestyle='dashed', linewidth=1)
  #df.plot.kde()
  plt.title('HISTOGRAM')
  plt.xlabel('Values')
  plt.ylabel('Counter')
  plt.xticks(rotation = 45)
  plt.yticks(rotation = 360)
  plt.minorticks_on()
  plt.grid(which = 'minor', ls = '--')
  plt.grid(which = 'major', linewidth = 0.5)

def describe_columns(df : pd.Series):

  print(f'Количество пропусков : {df.isna().mean()}')
  print(f'Тип данных - {df.dtype}')
```
